# Linked List

```python
class Node:
    
    __splits__ = 'content', 'next'
    def __init__(self, content):
        self.content = content
        self.next    = None
        
class LinkedList:
    
    def __init__(self):
        self.head = None
        self.tail = None 
        self.size = 0
        
    def add_first(self,e):
        self.head  = Node(e, self.head) 
        self.size +=1
        
    def delete_first(self):
        if self.head:
            self.head  = self.head.next
            self.size -= 1
        else: 
            raise Empty('Linked list is empty')
        
    def add_last(self, e):
        if self.tail: 
            self.tail.next = Node(e)
            self.tail      = self.tail.next
            self.size     += 1
        else: 
            self.tail = Node(e)
            self.size     += 1
    
```

# Linked list implementation of stack

```python
class LinkedStack: 
    
    class _Node:
        __splits__ = '_content'  , '_next'
        def __init__(self, content, next):
            self._content = content
            self._next    = None 
    
    def __init__(self):
        self._head = None
        self._size = 0 
    
    def __len__(self):
        return self._size
    
    def is_empty(self):
        return self.size == 0
    
    def top(self):
        if self.is_empty():
            raise Empty('Stack is empty')
        return self._head._content
        
    def push(self,e):
        self._head = self._Node(e, self._head)
        self._size +=1

    def pop(self):
        if self.is_empty():
            raise Empty('Stack is empty')
        answer = self._head._content 
        self._head  = self._head._next
        self._size -= 1
        return answer
            
        
```

# Linked list implementation of queue

```python
class LinkedQueue:
    
    class _Node:
        
        __splits__ = '_content' , '_next'
        def __init__(self, content, next):
            self._content = content 
            self._next    = next
    
    def __init__(self):
        self._head = None
        self._tail = None 
        self._size = 0
    
    def __len__(self):
        return self._size
    
    def is_empty(self):
        return self._size==0
    
    def first(self):
        if is_empty():
            raise Empty('Queue is empty')
        else:
            return self._head._content
        
    def dequeue(self):
        if self.is_empty():
            raise Empty('Queue is empty')
        answer = self._head
        self._head  = self._head._next
        self._size -= 1
        
        if self.is_empty():
            self._tail = None   # Tail has to be explicitly set to None if the removed head was the tail 
        
        return answer 
        
    def enqueue(self, e):
        new_node = self._Node(e, None)
        if is_empty():
            self._head  =  new_node
        else: 
            self._tail._next = new_node
        
        self._tail = new_node
        self._size += 1
    
```

# Circularly linked list 

```python
class CircularQueue:
    
    class _Node:
        __splits__ = '_content' , '_next'
        def __init__(self, content, next):
            self._content = content
            self._next    = next
    
    def __init__(self):
        self._tail = None
        self._size = None
        
    def __len__(self):
        return self._size
    
    def is_empty(self):
        return self._size == 0
    
    def first(self):
        if is_empty():
            raise Empty('Queue is empty')
        return self._tail._next._content 
            
    def dequeue(self):
        if is_empty():
            raise Empty('Queue is empty')
        
        old_head = self._tail._next
        if self._size == 1:
            self._tail = None 
        else: 
            self._tail._next = old_head._next
        self._size -= 1
        return old._head._next
    
    def enqueue(self, e):
        
        new_element = self._Node(e, None)
        if is_empty():
            new_element._next = new_element
        else:
            new_element._next  = self._tail._next
            self._tail._next = new_element 
            
        self._tail  = new_element
        self._size += 1
        
    def rotate(self):
        if self._size > 0:
            self._tail = self._tail._next
        
        
        
```

# Doubly Linked List : The base
    - Uses dummy gaurd nodes (sentinels)

```python
class _DoublyLikedBase:
    
    class _Node:
        
        __splits__ = '_content' , '_prev' , '_next'
        def __init__(self , content, prev, next):
            self._content = content
            self._prev    = prev
            self._next    = next 
            
    def __init__(self):
        self._size = 0 
        self._header  = self._Node(None, None, None)
        self._trailer = self._Node(None, None, None)
        
        self._header._next  = self._trailer
        self._trailer._next = self._header 

    def __len__(self):
        return self._size 
    
    def is_empty():
        return self._size == 0
    
    def _insert(self, e , pred, succ):
        node = self._Node(e, pred, succ)
        pred._next = node 
        succ._prev = node
        self._size +=1
        return node 
    
    def _delete(self, node):
        pred = node._prev
        succ = node._next
        pred._next  = succ
        succ._prev  = pred
        self._size -= 1
        content = node._content
        node._prev = node._next = node._content = None 
        return content
        
        
        
```

# Linked Deque

```python
class LinkedDeque(_DoublyLikedBase):
    
    def first(self):
        if self.is_empty():
            raise Empty('Deque is empty')
        return self._header._next._content 
    
    def last(self):
        if self.is_empty():
            raise Empty('Deque is empty')
        return self._trailer._prev._content   
    
    def insert_first(self, e):
        self._insert(e, self._header, self._header._next)
        
    def insert_last(self, e):
        self._insert(e, self._trailer._prev, self._trailer)
        
    def delete_first(self):
        if self.is_empty():
            raise Empty('Deque is empty')
        return self._delete(self._header._next)
        
    def delete_last(self):
        if self.is_empty():
            raise Empty('Deque is empty')
        return self._delete(self._trailer._prev)
    
```

# Positional list

```python
class PositionalList(_DoublyLikedBase):
    
    class Position:
        
        def __init__(self, container, node):
            self._container = container  
            self._node      = node 
        
        def content(self):
            return self._node_.content
        
        def __eq__(self, other):
            return type(other) is type(self) and other._node is self._node
        
        def __ne__(self, other):
            return not (self == other)
            
    
    def _validate(self, p):
        if not isinstance(p , self.Position):
            return TypeError('p must be proper postional type !')
        if p._container is not self: 
            raise ValueError('p does not belong to this container')
        if p._node._next is None:
            raise ValueError('p is no longer valid')
        return p._node 
            
    def _make_position(self, node):
        if node is self._header or node is self._trailer:
            return None 
        else: 
            return self.Position(self, node)
        
    def first(self):
        return self._make_position(self._header._next)
    
    def last(self):
        return self._make_position(self._trailer._prev)
    
    def before(self, p):
        node = self._validate(p)
        return self._make_position(node._prev)

    def after(self, p):
        node = self._validate(p)
        return self._make_position(node._next)
    
    def __iter__(self):
        cursor = self.first()
        while cursor is not None:
            yield cursor.content()
            cursor = self.after(cursor)
    
    def _insert(self, e, pred, succ):
        node.super()._insert(e, pred, succ)
        return self._make_position(node)
    
    def add_first(self):
        return self._insert(e, self._header, self._header._next)
    
    def add_last(self):
        return self._insert(e, self._trailer._prev, self._trailer)
    
    
    def add_before(self, p, e):
        original = self._validate(p)
        return self._insert(e, original._prev, original)
    
    def add_after(self, p, e):
        original = self._validate(p)
        return self._insert(e, original, original._next)
    
    def delete(self, p):
        original = self._validate(p)
        return self._delete(original)

    def replace(self, p, e):
        pos  = self._validate(p)
        old_content = pos._content
        pos._content  = e
        return old_content
```
