```python
def countAndSay(n):
    
    if n==1:
        return '1'
    else:
        
        current_str = '1'
        for i in range(1,n):
            res = ''
            cnt = 1
            prev_dig = current_str[0]
            for j in range(1, len(current_str)):
                if current_str[j] == prev_dig:
                    cnt +=1
                else:
                    res += str(cnt) + prev_dig
                    prev_dig = current_str[j]
                    cnt = 1
            res +=  str(cnt) + prev_dig
            current_str = res
            
        return res

                

    
```
